Write a program that converts a binary (base 2) number to decimal 
(base 10). Your program should begin by reading the binary number 
from the user as a string. Then it should compute the equivalent decimal 
number by processing each digit in the binary number. Finally, your 
program should display the equivalent decimal number with an 
appropriate message.

```python
binary = input("Enter a number  ")

decimal = 0
for i, digit in enumerate(reversed(binary)):
    if digit == '1':
        decimal += 2**i
    elif digit != '0':
        print(f"{digit}' is not a valid binary digit.")
        exit()
print(f"The decimal equivalent of binary {binary} is {decimal}.")
```
