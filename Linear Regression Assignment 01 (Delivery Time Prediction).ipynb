```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('C:\\Mohit -Cohort138\\ML\Assignment\\Project Assignment 1 and 2 - Simple Linear Regression\\delivery_time.csv')
data.head(5)
```

```python
print("Is there any null values")
data.isna().sum()
```

```python
data.dtypes
```

```python
Delivery=data['Delivery Time']
Sorting=data['Sorting Time']
```

```python
data.corr()
```

```python
sns.heatmap(data.corr(),annot=True,cmap='RdBu')
```

```python
sns.boxplot(x = data['Delivery Time'])
```

```python
from sklearn.linear_model import LinearRegression 
model = LinearRegression()
```

```python
model
```

```python

Delivery=data[['Delivery Time']]
Sorting=data[['Sorting Time']]
print("Model Training ---- ")
model_train=model.fit(Delivery,Sorting,)
```

```python
m=model_train.coef_
m
```

```python
c=model_train.intercept_
c
```

```python
x=35
y=m*x+c
y
```

```python
plt.scatter(data['Delivery Time'],data['Sorting Time'],label='Actual Data')
plt.plot(data['Delivery Time'],model_train.predict(Delivery),color='yellow',label='Regression line')
plt.title("Delivery Time Prediction Model")
plt.legend()
plt.show()
```

```python

acc=model.score(Delivery,Sorting)*100
print("Model Evalution r2 is :", acc)
```

```python
err=100-acc
err
print("Error is :", err)
```

```python
x=[15.6,23.6,10.6]
y=m*x+c
y
```

```python

df = pd.DataFrame({ 'DeliveryTime':x,
                  'SortingTime':y.flatten()}
                 )
df
```

```python
df.to_csv('Delivery Time Predict Results.csv',index=False )
print("Export Complete. Predicted Results:")
print(df.head(3))
```
