```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model
```

```python
print("Model Training in process ....")
YrExp=data1[['YearsExperience']]
Salary=data1[['Salary']]
model_train=model.fit(YrExp,Salary)
model_train
```

```python
m=model_train.coef_
c=model_train.intercept_
```

```python
x=15
#y=mx+c
y=m*x+c
```

```python
x=[10,12,16]
y=m*x+c
y.reshape(-1,1)
y
```

```python
plt.scatter(data1['YearsExperience'],data1['Salary'],label='Actual Data')
plt.plot(data1['YearsExperience'],model_train.predict(YrExp),color='grey',label='Regression line')
plt.title("Salary Prediction Model")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score
acc1=model_train.score(YrExp,Salary)*100
print("Model Evalution of Salary Prediction r2 is :", acc1)
```

```python
err1=100-acc1
print("Error of Salary Prediction r2 is :", err1)
```
